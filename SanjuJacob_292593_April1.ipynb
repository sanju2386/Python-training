```python
import random

class Car:
    def __init__(self, brand, model, year, engine_capacity, mileage=0, fuel_level=50):
       # intialization
        self.brand = brand
        self.model = model
        self.year = year
        self.engine_capacity = " ".join([str(engine_capacity), "cc"])
        self.mileage = mileage
        self.fuel_level = fuel_level  # Fuel level in liters

    def get_mileage_per_liter(self):
        #Returns mileage per liter based on engine capacity.
        if "999cc" <= self.engine_capacity <= "1000cc":
            return random.uniform(25, 35)  # Random value between 25-35 km/l
        elif "1200cc" <= self.engine_capacity <= "1500cc":
            return random.uniform(17, 27)  # Random value between 17-27 km/l
        elif "2000cc" <= self.engine_capacity <= "3000":
            return random.uniform(12, 25)  # Random value between 12-25 km/l
        else:
            return 15  # Default value if engine capacity doesn't match

    def drive(self, distance):
        #Simulates driving the car and updates mileage and fuel level.
        mileage_per_liter = self.get_mileage_per_liter()
        max_distance = self.fuel_level * mileage_per_liter  # Maximum possible distance with current fuel

        if distance <= max_distance:
            self.mileage += distance
            self.fuel_level -= distance / mileage_per_liter
            print(f"Drove {distance} km. New mileage: {self.mileage} km, Fuel level: {self.fuel_level:.2f} L")
        else:
            print(f"Not enough fuel! You can only drive {max_distance:.2f} km with the current fuel.")

    def refuel(self, amount):
        # Refuels the car by the given amount, ensuring the fuel level does not exceed 50 liters."""
        self.fuel_level = min(50, self.fuel_level + amount)  # Max fuel level set to 50 liters
        print(f"Refueled {amount} L. Current fuel level: {self.fuel_level} L")

    def display_info(self):
        """Displays the car's details."""
        print(f"Car Details:\nBrand: {self.brand}\nModel: {self.model}\nYear: {self.year}\nEngine: {self.engine_capacity}cc\nMileage: {self.mileage} km\nFuel Level: {self.fuel_level:.2f} L")

# Example Usage:
car1 = Car("Toyota", "Corolla", 2022, engine_capacity=1500, mileage=20000, fuel_level=30)
car1.display_info()
car1.drive(200)  # Try driving 200 km
car1.refuel(10)  # Refuel 10 liters
car1.display_info()
car1.drive(500)  # Try driving 500 km with the available fuel
```

```python
class Car:
    def __init__(self, brand, model, year, engine_capacity, mileage, fuel_level):
        self.brand = brand
        self.model = model
        self.year = year
        self.engine_capacity = engine_capacity 
        self.mileage = mileage  # Current mileage in km
        self.fuel_level = fuel_level  # Fuel level in liters

    def get_mileage_per_liter(self):
        # Define realistic mileage per liter based on engine capacity
        if   600<= self.engine_capacity <= 999:
            return 35
        elif 999 <= self.engine_capacity <= 1000:
            return 25  # km per liter for engines 999cc to 1000cc
        elif 1200 <= self.engine_capacity <= 1500:
            return 17  # km per liter for 1200cc to 1500cc engines
        elif 2000 <= self.engine_capacity <= 3000:
            return 12  # km per liter for 2000cc to 3000cc engines
        else:
            return 10  # km per liter for any larger engine

    def drive(self, distance):
        # Get the mileage per liter for the current engine size
        mileage_per_liter = self.get_mileage_per_liter()

        # Calculate how much fuel will be used for the distance
        fuel_needed = distance / mileage_per_liter

        if self.fuel_level >= fuel_needed:
            # Update the car's mileage and fuel level
            self.mileage += distance
            self.fuel_level -= fuel_needed
            print(f"Drove {distance} km. New mileage: {self.mileage} km, Fuel level: {self.fuel_level:.2f} L")
        else:
            print("Not enough fuel! Please refuel.")

    def refuel(self, amount):
        self.fuel_level += amount
        print(f"Refueled {amount} L. Current fuel level: {self.fuel_level:.2f} L")

    def display_info(self):
        print("Car Details:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Engine: {self.engine_capacity}")
        print(f"Mileage: {self.mileage} km")
        print(f"Fuel Level: {self.fuel_level:.2f} L")


# Example usage
car = Car("Toyota", "Corolla", 2022, 1500, 20000, 30.00)  # Engine 1500cc, 20,000 km mileage, 30 L fuel
car.display_info()

car.drive(200)  # Drive 200 km
car.refuel(10)  # Refuel with 10 L
car.display_info()

car.drive(500)  # Drive 500 km
car.display_info()
```

```python
class BankAccount:
    def __init__(self, balance, account_number, min_balance):
        self.balance = balance
        self.account_number = account_number
        self.min_balance = min_balance

    def deposit(self, account_number, deposit):
        if self.account_number == account_number:
            self.balance += deposit
            return f"Deposit successful! New balance: {self.balance}."
        else:
            return "Enter valid account number."

    def withdraw(self, account_number, withdrawal):
        if self.account_number == account_number:
            # Check if there are sufficient funds to withdraw and after withdrawal, 
            # balance should be greater than or equal to the minimum balance.
            if self.balance > withdrawal and (self.balance - withdrawal) >= self.min_balance:
                self.balance -= withdrawal
                return f"Withdrawal successful! New balance: {self.balance}."
            else:
                if self.balance <= withdrawal:
                    return "Insufficient funds."
                else:
                    return f"Insufficient funds. Balance after withdrawal should not be less than {self.min_balance}."
        else:
            return "Enter valid account number."

    def get_balance(self):
        return f"Current balance: {self.balance}"

    # def display_info(self): 
    #    print("current balance {}".format(self.get_balance()))
    #    print("Breadth is {}".format(self.breadth))
    #    #print("Area is {}".format(self.a))
    #    print("Area is {}".format(self.area()))
    #    print("Perimeter is {}".format(self.perimeter()))
    #    print("Perimeter is {}".format(self.perimeter()))
    #    print("Is object is {}".format(self.is_square()))
       

# Example usage:
account = BankAccount(balance=5000, account_number="123456", min_balance=1000)

print(account.deposit("123456", 1500))  # Deposit
print(account.withdraw("123456", 4000))  # Withdraw (insufficient balance after withdrawal)
print(account.withdraw("123456", 2500))  # Withdraw (sufficient funds)
print(account.get_balance())  # Get balance
```

```python
# ElectricCar Class (Child Class)
class ElectricCar(Car):
    def __init__(self, brand, model, year, engine_capacity, mileage, battery_capacity, charge_level=100):
        # Call the parent class constructor with default fuel level = 0 (since EVs don't use fuel)
        super().__init__(brand, model, year, engine_capacity, mileage, fuel_level=0)
        
        # EV-specific attributes
        self.battery_capacity = battery_capacity  # Battery capacity in kWh
        self.charge_level = charge_level  # Battery charge level in %

    def charge(self, amount):
        #Charges the battery but ensures it doesn't exceed 100%
        self.charge_level = min(self.charge_level + amount, 100)
        print(f"Charged {amount}%. Battery level: {self.charge_level}%")

    def display_info(self):
        # Override display_info() to include electric car details
        super().display_info()  # Call parent method to display general car info
        print(f"Battery Capacity: {self.battery_capacity} kWh")
        print(f"Charge Level: {self.charge_level}%")
        print("-" * 40)
```
