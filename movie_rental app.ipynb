```python
class Movie:
    def __init__(self, title, genre, total_copies, stock, rented_count):
        self.title = title              # Title of the movie (e.g. "Dune")
        self.genre = genre              # Genre of the movie (e.g. "Sci-Fi")
        self.total_copies = total_copies  # Total number of copies the store owns
        self.stock = stock              # Currently available copies for rent
        self.rented_count = rented_count  # Number of times the movie has been rented out

    # Method to check if the movie is available for rent
    def is_available(self):
        return self.stock > 0  # If stock > 0, it means the movie is available

    # Method to rent the movie, reducing stock by 1
    def rent_movie(self):
        if self.is_available():  # Check if the movie is available
            self.stock -= 1
            self.rented_count += 1  # Increment the rented count
            print(f"{self.title} has been rented out.")
        else:
            print(f"{self.title} is currently out of stock.")

    # Method to return the movie, increasing stock by 1
    def return_movie(self):
        if self.rented_count < self.total_copies:  # Ensure there is a rented copy to return
            self.stock += 1  # Increment stock by 1 when a movie is returned
            print(f"{self.title} has been returned.")
        else:
            print(f"No rented copies of {self.title} to return.")
```

```python
# Creating an instance of the Movie class
movie = Movie(title="Dune", genre="Sci-Fi", total_copies=3, stock=2, rented_count=1)

# Display initial values
print(f"Initial stock for '{movie.title}': {movie.stock}")
print(f"Initial rented count for '{movie.title}': {movie.rented_count}")

# Rent a movie
print("\nRenting a movie...")
movie.rent_movie()

# Display values after renting
print(f"Stock after renting: {movie.stock}")
print(f"Rented count after renting: {movie.rented_count}")

# Return a movie
print("\nReturning a movie...")
movie.return_movie()

# Display values after returning
print(f"Stock after returning: {movie.stock}")
print(f"Rented count after returning: {movie.rented_count}")
```

```python
from abc import ABC, abstractmethod
from datetime import datetime
import random

# Movie Class
class Movie:
    def __init__(self, title, genre, total_copies, stock, rented_count):
        self.title = title              # Title of the movie (e.g. "Dune")
        self.genre = genre              # Genre of the movie (e.g. "Sci-Fi")
        self.total_copies = total_copies  # Total number of copies the store owns
        self.stock = stock              # Currently available copies for rent
        self.rented_count = rented_count  # Number of times the movie has been rented out

    # Method to check if the movie is available for rent
    def is_available(self):
        return self.stock > 0  # If stock > 0, it means the movie is available

    # Method to rent the movie, reducing stock by 1
    def rent_movie(self):
        if self.is_available():  # Check if the movie is available
            self.stock -= 1
            self.rented_count += 1  # Increment the rented count
            print(f"{self.title} has been rented out.")
        else:
            print(f"{self.title} is currently out of stock.")

    # Method to return the movie, increasing stock by 1
    def return_movie(self):
        if self.rented_count < self.total_copies:  # Ensure there is a rented copy to return
            self.stock += 1  # Increment stock by 1 when a movie is returned
            print(f"{self.title} has been returned.")
        else:
            print(f"No rented copies of {self.title} to return.")


# Abstract User Class
class User(ABC):
    user_id_set = set()  # This will store the generated user IDs
    
    def __init__(self, name):
        self.name = name
        self.user_id = self.generate_unique_id()  # Automatically generate a unique ID
        self.rented_movies = []  # List to store rented movies as dicts

    @abstractmethod
    def get_role(self):
        pass

    def rent_movie(self, movie, rent_date, due_date):
        """Rent a movie and add it to the rented_movies list"""
        rental_info = {
            "title": movie.title,
            "rent_date": rent_date,
            "due_date": due_date
        }
        self.rented_movies.append(rental_info)
        print(f"{self.name} rented the movie '{movie.title}'.")

    def return_movie(self, movie_title):
        """Return a movie and remove it from the rented_movies list"""
        for rental in self.rented_movies:
            if rental["title"] == movie_title:
                self.rented_movies.remove(rental)
                print(f"{self.name} returned the movie '{movie_title}'.")
                return
        print(f"{movie_title} is not found in {self.name}'s rented movies.")

    @classmethod
    def generate_unique_id(cls):
        """Generate a unique user ID"""
        while True:
            new_id = random.randint(1000, 9999)
            if new_id not in cls.user_id_set:
                cls.user_id_set.add(new_id)  # Add the ID to the set
                return str(new_id)  # Return the ID as a string


# Member Class (concrete implementation of User)
class Member(User):
    def __init__(self, name):
        super().__init__(name)  # The `user_id` is automatically generated in the User class constructor

    def get_role(self):
        return "Member"

    def get_rental_history(self):
        # Return rental history as an iterator
        for rental in self.rented_movies:
            yield rental  # Yield each rental record as an iterator


# Example usage:

# Create a Movie
dune = Movie("Dune", "Sci-Fi", total_copies=3, stock=3, rented_count=0)

# Create a Member (user_id will be generated automatically)
member1 = Member("Sanju")

# Rent the movie
member1.rent_movie(dune, rent_date=datetime.now(), due_date=datetime.now())

# Check inventory
print(f"Remaining stock for {dune.title}: {dune.stock}")

# Return the movie
member1.return_movie(dune.title)

# Check inventory again
print(f"Stock after return: {dune.stock}")
```
