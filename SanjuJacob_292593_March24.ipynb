```python
def second_largest(nums):
    if len(nums) < 2:
        return "List must have at least two numbers"
    
    first, second = float('-inf'), float('-inf')  # Initialize variables
    
    for num in nums:
        if num > first:  # Found new largest
            second, first = first, num
        elif first > num > second:  # Found second largest
            second = num
    
    return second if second != float('-inf') else "No second largest number"

# Example usage
print(second_largest([34, 56, 67, 43]))  # Output: 56
print(second_largest([10, 10, 10]))  # Output: No second largest number
```

```python
def recjoin(l):
    if len(l) == 0:
        return ""  # ✅ Return an empty string instead of None

    return str(l[0]) + (" " + recjoin(l[1:]) if len(l) > 1 else "")  # ✅ Handle space properly

# Example Usage
print(recjoin(["hello", "world", "!"]))  # Output: "hello world !"
print(recjoin(["Python", "is", "awesome"]))  # Output: "Python is awesome"
print(recjoin([]))  # Output: ""
```

# Q1:Write a recursive function to compute the nth Fibonacci number.

```python
def fibonacci(n):
    a, b = 0, 1  # Initialize first two Fibonacci numbers
    fib_series = []  # List to store Fibonacci sequence

    for _ in range(n):
        fib_series.append(a)  # ✅ Append the current Fibonacci number (not a + b)
        a, b = b, a + b  # ✅ Update values for the next iteration

    return fib_series  # Return the generated sequence

        
```

```python
fibonacci(9)
```

```python
def fibonacci(n):
    if n < 0:
        return "Input must be a non-negative integer"
    if n == 0:
        return 0  # Base case: Fib(0) = 0
    if n == 1:
        return 1  # Base case: Fib(1) = 1
    return fibonacci(n - 1) + fibonacci(n - 2)  # Recursive case

# Example Usage
print(fibonacci(9))  # Output: 8
```

```python
def find_common(*lists):
    if not lists:
        return []  # Return empty list if no input lists are given
    
    common_elements = set(lists[0])  # Start with the first list as a set
    
    for lst in lists[1:]:  # Iterate through remaining lists
        common_elements.intersection_update(lst)  # Keep only common elements
    
    return list(common_elements)  # Convert set back to list

# Example Usage
list1 = [1, 2, 3, 4, 5]
list2 = [2, 3, 5, 7, 9]
list3 = [1, 2, 5, 8, 10]

print(find_common(list1, list2, list3))  # Output: [2, 5]
```
